```python
import pandas as pd
train_df = pd.read_csv("/home/rnanawa1/GeoLifeLearn/data/species25/train/metadata.csv")
test_df = pd.read_csv("/home/rnanawa1/GeoLifeLearn/data/species25/test/metadata.csv")
val_df = pd.read_csv("/home/rnanawa1/GeoLifeLearn/data/species25/val/metadata.csv")
```

```python
# get list of unique values in "label" column in train_df
unique_labels = train_df['label'].unique()
print(unique_labels)

# convert labels to integers
label_to_int = {label: i for i, label in enumerate(unique_labels)}

```

```python
train_df['label'] = train_df['label'].apply(lambda x: label_to_int[x])
test_df['label'] = test_df['label'].apply(lambda x: label_to_int[x])
val_df['label'] = val_df['label'].apply(lambda x: label_to_int[x])
```
