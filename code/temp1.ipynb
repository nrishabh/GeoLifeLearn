```python
import pandas as pd
```

```python
def generate_img_path(surveyId):
    surveyId = str(surveyId)
    if len(surveyId) < 3:
        return "po_train_patches_rgb/" + surveyId + "/" + surveyId + ".jpeg"
    elif len(surveyId) == 3:
        return "po_train_patches_rgb/" + surveyId[-2:] + "/" + surveyId[-3] + "/" + surveyId + ".jpeg"
    else:
        return "po_train_patches_rgb/" + surveyId[-2:] + "/" + surveyId[-4:-2] + "/" + surveyId + ".jpeg"
```

```python
df = pd.read_csv("/home/rnanawa1/GeoLifeLearn/output/PresenceOnlyOccurrences/GLC24-PO-metadata-train.csv")
```

```python
# group by speciesId and get counts for each speciesId
species_counts = df.groupby('speciesId').size()

# arrange by descending order of counts
species_counts = species_counts.sort_values(ascending=False)

# get top 500 speciesId
top_500_species = species_counts.index[:25]

# filter the dataframe to only include the top 500 species
df_top_500 = df[df['speciesId'].isin(top_500_species)].copy()

df_top_500['img_path'] = df_top_500['surveyId'].apply(generate_img_path)

# keep only "speciesId" and "surveyId" columns. Remove the rest.
df_top_500 = df_top_500[['speciesId', 'img_path']]
```

```python
df_top_500.to_csv("shortlist.csv", index=False)
# df = pd.read_csv("shortlist.csv")
```
